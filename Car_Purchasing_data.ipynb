```python
from sklearn.linear_model import LinearRegression   #Give the data to the model
 
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
reg.coef_ # m Value
```

```python
# c value

reg.intercept_
```

```python
#train_accuracy

y_train_pred = reg.predict(X_train)
```

```python
from sklearn.metrics import r2_score
```

```python
print(f'train accuracy: {r2_score(y_train,y_train_pred)}')
```

```python
#Finding the loss

print(f'train loss is: {1-r2_score(y_train,y_train_pred)}')
```

```python
# Test accuracy

y_test_pred = reg.predict(X_test)
```

```python
print(f'test accuracy: {r2_score(y_test,y_test_pred)}')
```

```python
print(f'test loss: {1-r2_score(y_test,y_test_pred)}')
```

```python
# Confusion Matrix

from sklearn.metrics import confusion_matrix
confusion_matrix(y_train,y_train_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_train,y_train_pred))
```
